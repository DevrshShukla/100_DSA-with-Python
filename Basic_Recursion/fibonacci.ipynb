### fibonacci using recurison


```python
def febonacci(n):
    if n <= 1:
        return n
    last = febonacci(n-1)
    s_last = febonacci(n-2)
    return last + s_last

print(febonacci(5))
```

| Call No. | Function Call | Returned Value | Reason          |
| -------- | ------------- | -------------- | --------------- |
| 1        | fib(5)        | 5              | fib(4) + fib(3) |
| 2        | fib(4)        | 3              | fib(3) + fib(2) |
| 3        | fib(3)        | 2              | fib(2) + fib(1) |
| 4        | fib(2)        | 1              | fib(1) + fib(0) |
| 5        | fib(1)        | 1              | Base case       |
| 6        | fib(0)        | 0              | Base case       |
| 7        | fib(1)        | 1              | Base case       |
| 8        | fib(2)        | 1              | fib(1) + fib(0) |
| 9        | fib(1)        | 1              | Base case       |
| 10       | fib(0)        | 0              | Base case       |

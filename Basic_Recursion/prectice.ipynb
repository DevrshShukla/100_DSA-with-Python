### Print Name 5 Times

```python

def print_name(i,n):
    if i > n:
        return
    print("Devarsh Shukla" , i)
    print_name(i+1,n)
print_name(1,6)
```

### Print 1 to N numbers

```python

def print_num(i,n):
    if i > n:
        return
    print(i)
    print_num(i+1,n)
print_num(1,15)
    
```

### Print N to 1 numebrs

```python
def reverse(i,n):
    if i < 1:
        return
    print(i)
    reverse(i-1,n)
reverse(15,1)
    
```

### Sum Of N numbers

```python
def sum_num(i,sum):
    if i < 1:
        print(sum)
        return
    sum_num( i - 1 , sum + i)
sum_num(3,0)
```

| Call No. | i | sum | Action Taken       | Next Call     |
| -------- | - | --- | ------------------ | ------------- |
| 1        | 3 | 0   | sum + i → 0 + 3    | sum_num(2, 3) |
| 2        | 2 | 3   | sum + i → 3 + 2    | sum_num(1, 5) |
| 3        | 1 | 5   | sum + i → 5 + 1    | sum_num(0, 6) |
| 4        | 0 | 6   | i < 1 → print(sum) | Stop          |


```python
# functional method to calculate sum of n numbers using recursion
def ans(n):
    if n == 0:
        return 0
    return n + ans(n-1)
print(ans(3))
```

| Call No. | Function Call | Returned Value | Reason     |
| -------- | ------------- | -------------- | ---------- |
| 1        | ans(3)        | 6              | 3 + ans(2) |
| 2        | ans(2)        | 3              | 2 + ans(1) |
| 3        | ans(1)        | 1              | 1 + ans(0) |
| 4        | ans(0)        | 0              | Base case  |


### Factorial Of N numbers

```python
# parametered function
def fact(i , f):
    if i == 0:
        print(f)
        return 1
    fact(i-1 , f * i)
fact(4 , 1)
```

```python
# functional method to calculate factorial of n using recursion
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)
print(factorial(4))
```

| Call No. | Function Call | Returned Value | Explanation      |
| -------- | ------------- | -------------- | ---------------- |
| 1        | factorial(4)  | 24             | 4 × factorial(3) |
| 2        | factorial(3)  | 6              | 3 × factorial(2) |
| 3        | factorial(2)  | 2              | 2 × factorial(1) |
| 4        | factorial(1)  | 1              | 1 × factorial(0) |
| 5        | factorial(0)  | 1              | Base case        |

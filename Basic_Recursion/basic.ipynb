```python
# Recursion example that leads to infinite recursion
def first():
    print("First function")
    first()

first()
```

```python
c = 0
def fun():
    global c # Use 'global' to modify the global variable 'c'
    if c == 3:
        print("Done")
        return
    print("Base Case not reached")
    c += 1
    fun()
fun()
```

```python
# print 1 to n numbes
def print_1_to_n(n):
    num = 1
    if num == n:
        print(num)
        return
    print(num)
    num += 1
    print_1_to_n(n)
    
# print_1_to_n(5)
    
```

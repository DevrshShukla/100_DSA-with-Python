### check pallindrom string

```python
def pallindrom( i , a ):
    n = len(a)
    if i > n // 2:
        return True
    if a[i] != a[n-1-i]:
        return False
    return pallindrom(i+1, a)
    
pallindrom(0, "madam")
```

| Step | i | a[i] | n-1-i | a[n-1-i] | Comparison | Result        |
| ---- | - | ---- | ----- | -------- | ---------- | ------------- |
| 1    | 0 | m    | 4     | m        | m == m     | Continue      |
| 2    | 1 | a    | 3     | a        | a == a     | Continue      |
| 3    | 2 | d    | 2     | d        | d == d     | Continue      |
| 4    | 3 | —    | —     | —        | i > n//2   | Return `True` |

### Swap Number using 2-pointer + recursion

```python
def swap_num( i , j , a ):
    if i > j:
        return 
    a[i] , a[j] = a[j], a[i]
    swap_num( i + 1 , j - 1 , a )
    
a = [1, 2, 3, 4, 5]
swap_num( 0 , len(a) - 1 , a )
print(a)
```


| Call | i | j | List State      |
| ---- | - | - | --------------- |
| 1    | 0 | 4 | [5, 2, 3, 4, 1] |
| 2    | 1 | 3 | [5, 4, 3, 2, 1] |
| 3    | 2 | 2 | [5, 4, 3, 2, 1] |
| 4    | 3 | 1 | Stop            |


### Swap Number using 1-pointer + Recursion

```python
def swap_n( i , n , a):
    if  i > n // 2:
        return
    a[i] , a[n-i-1] = a[n-i-1] , a[i]
    swap_n( i+1 , n , a)
    
a = [1, 2, 3, 4, 5]
swap_n( 0 , len(a) , a )
print(a)
```

| Call No. | i | Left Element `a[i]` | Right Index `n-i-1` | Right Element `a[n-i-1]` | After Swap      | Next Call     |
| -------- | - | ------------------- | ------------------- | ------------------------ | --------------- | ------------- |
| 1        | 0 | 1                   | 4                   | 5                        | [5, 2, 3, 4, 1] | swap_n(1,5,a) |
| 2        | 1 | 2                   | 3                   | 4                        | [5, 4, 3, 2, 1] | swap_n(2,5,a) |
| 3        | 2 | 3                   | 2                   | 3                        | [5, 4, 3, 2, 1] | swap_n(3,5,a) |
| 4        | 3 | —                   | —                   | —                        | i > n//2 → Stop | Return        |


<br>


| Type  | Complexity                 |
| ----- | -------------------------- |
| Time  | **O(n)**                   |
| Space | **O(n)** (recursive stack) |

### String hashing

```python
s = [ "a" , "b" , "c" , "c" , "b" , "b" ]
m = [ "a" , "b" , "c" , "d" , "e" , "f" ]
dict = {}

for i in s:
    if i in dict:
        dict[i] += 1
    else:
        dict[i] = 1
dict
```

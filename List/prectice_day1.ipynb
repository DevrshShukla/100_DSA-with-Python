```
# Sum of Elements
arr = [1,2,3,4,5]
sum = 0
for i in arr:
    sum += i
print(sum)
```

```
# Find Maximum & Minimum
arr = [15,20,45,80,56]
max_num = arr[0]
min_num = arr[1]

for i in range(len(arr)):
      if arr[i] > max_num:
          max_num = arr[i]
      elif arr[i] < min_num:
          min_num = arr[i]
print("Max Number : " , max_num)
print("Min Number : " , min_num)
```

```
# Count Even & Odd
arr = [1,2,3,4,5]
odd = 0
even = 0

for i in arr:
  if i % 2 == 0:
    even += 1
  else:
    odd += 1
print("Even : " , even)
print("Odd : " , odd)
```

```
# Reverse a List
arr = [1,2,3,4,5,6,58,20]
i = 0
j = len(arr) - 1

while i < j:
  tmp = arr[i]
  arr[i] = arr[j]
  arr[j] = tmp
  i+=1
  j-=1
print(arr)
```

```
# Remove Duplicates
# [1, 2, 2, 3, 4, 4, 5]

# Output:
# [1, 2, 3, 4, 5]

a =  [0,0,8,1, 2, 2, 3, 4, 4, 5]
i = 0
j = 1
# print(len(a))
while i < len(a) - 1:
  if a[i] != a[i+1]:
    a[j] = a[i+1]
    j+=1
  i+=1
print(j)
print(a[:j])
```

```
# Second Largest Element
arr = [1,2,3,4,5,6,58,20]
max_val = max( arr[0] , arr[1])
Smax = min( arr[0] , arr[1])

for i in range( 2 , len(arr) ):
  if arr[i] > max_val:
    Smax = max_val
    max_val = arr[i]
  else:
    if arr[i] > Smax:
      Smax = arr[i]

print(max_val)
print(Smax)

```

```python
# pair sum -> Brute Force

a = [2,7,11,15]
target = 13

for i in range(len(a)):
    for j in range(i+1 , len(a)):
        if a[i] + a[j] == target:
            print(i,j)  
# Output: 0 1
```

```python
# pair sum ->  optimal
a = [2,7,11,15]
target = 9

i = 0
j = len(a) - 1

while i < j:
    pair_sum = a[i] + a[j]
    if pair_sum == target:
        print(i,j)
        break
    elif pair_sum < target:
        i+=1
    else:
        j-=1
# Output: 0 1
         
     
```
